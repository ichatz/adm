```python
import networkx as nx

# Matplotlib is the default package for
# rendering the graphs
import matplotlib.pyplot as plt

#create an empty graph
G = nx.Graph()

#add three edges
G.add_edge('A','B');
G.add_edge('B','C');
G.add_edge('C','A');

#draw the graph
nx.draw(G)
plt.savefig("graph-1.png")
```

```python
plt.clf()
```

```python
G.remove_node('A')

nx.draw(G)
plt.savefig("graph-2.png")
```
